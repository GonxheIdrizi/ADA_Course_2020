## Data from World Happiness Report

The World Happiness Report is an annual publication of the United Nations Sustainable Development Solutions Network. It contains articles, and rankings of national happiness based on respondent ratings of their own lives, which the report also correlates with various life factors.

In this notebook we will explore the happiness of different countries and the features associated.
The datasets that we will use are available in *Data*: **happiness2020.pkl** and **countries_info.csv**.

Although the features are self-explanatory, here a summary: 

**happiness2020.pkl**
* country: *Name of the country*
* happiness_score: *Happiness score*
* social_support: *Social support (mitigation the effects of inequality)*
* healthy_life_expectancy: *Healthy Life Expectancy*
* freedom_of_choices: *Freedom to make life choices*
* generosity: *Generosity (charity, volunteers)*
* perception_of_corruption: *Corruption Perception*
* world_region: *Area of the world of the country*

**countries_info.csv**
* country_name: *Name of the country*
* area: *Area in sq mi*
* population: *Number of people*
* literacy: *Literacy percentage*

```python
import pandas as pd
%matplotlib inline

DATA_FOLDER = 'Data/'

HAPPINESS_DATASET = DATA_FOLDER+"happiness2020.pkl"
COUNTRIES_DATASET = DATA_FOLDER+"countries_info.csv"
```

## Task 1: Load the data

Load the 2 datasets in Pandas dataframes (called *happiness* and *countries*), and show the first rows.


**Hint**: Use the correct reader and verify the data has the expected format.

```python
# Write your code here
happiness = pd.read_pickle(HAPPINESS_DATASET)
happiness.country = happiness.country.apply(lambda a: a.lower())
happiness
```

```python
countries = pd.read_csv(COUNTRIES_DATASET)
countries
```

## Task 2: Let's merge the data

Create a dataframe called *country_features* by merging *happiness* and *countries*. A row of this dataframe must describe all the features that we have about a country.

**Hint**: Verify to have all the rows in the final dataframe

```python
# Write your code here
country_features = pd.merge(happiness,countries, left_on='country', right_on='country_name').drop(['country_name'],axis=1)
country_features
```

## Task 3: Where do people are happier?

Print the top 10 countries based on their happiness score (high is better).

```python
# Write your code here
country_features.nlargest(10,'happiness_score')['country']
```

We are interested to know in what world region the people are happier. 

Create and print a dataframe with the (1) average happiness score and (2) the number of contries for each world region.
Sort the result to show the happiness ranking.

```python
# Write your code here
group_avg_score = country_features.groupby(['world_region'])['happiness_score'].mean()
```

```python
group_num_countries = country_features.groupby(['world_region']).size()
group_num_countries.name = 'num_countries'
```

```python
merged = country_features.merge(group_avg_score, left_on='world_region', right_on='world_region')
merged = merged.merge(group_num_countries, left_on='world_region', right_on='world_region')
merged = merged.sort_values(by=['happiness_score_y'], ascending=False)
merged
```

The first region has only a few countries! What are them and what is their score?

```python
# Write your code here

region = merged[merged.world_region == merged.iloc[0]['world_region']][['country', 'happiness_score_x']]
region
```

## Task 4: How literate is the world?

Print the name of countries with a level of literacy of 100%. 

For each country, print the name and the world region with the format: *{region name} - {country name} ({happiness score})*

```python
country_features['literacy'] = country_features['literacy'].replace(',','.', regex=True).astype(float)
```

```python
countries_100 = country_features[country_features['literacy']==100]
countries_100
```

```python
# Write your code here
for country in countries_100['country'].unique():
    a = country_features[country_features['country']==country].happiness_score.item()
    b = country_features[country_features['country']==country].world_region.item()
    print(f" {{{b}}} - {{{country}}} ({{{a}}})")
```

What is the global average?

```python
# Write your code here
countries_100['happiness_score'].mean()
```

Calculate the proportion of countries with a literacy level below 50%. Print the value in percentage, formatted with 2 decimals.

```python
# Write your code here
countries_50 = country_features[country_features['literacy']<50]
percentage = len(countries_50)/float(len(country_features))*100
print(f' There is {percentage:.2f}% of countries with a literacy level below 50%')
```

Print the raw number and the percentage of world population that is illiterate.

```python
# Write your code here
popul_illiterate = countries_50['population'].sum()
total_population = country_features['population'].sum()
percentage = popul_illiterate/float(total_population)*100
print(f' There are {popul_illiterate} people, or {percentage:.2f}% of world population which is/are illiterate.')
```

## Task 5: Population density

Add to the dataframe a new field called *population_density* computed by dividing *population* by *area*.

```python
# Write your code here
country_features['population_density'] = country_features['population']/country_features['area']
country_features
```

What is the happiness score of the 3 countries with lowest population density?

```python
# Write your code here
country_features.nsmallest(3, 'population_density')[['country', 'happiness_score']]
```

## Task 6: Healty and happy?

Plot in scatter plot the happiness score (x) and healthy like expectancy (y).

```python
# Write your code here
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility

N = len(country_features)
x = country_features['happiness_score']
y = country_features['healthy_life_expectancy']
area = 50  # 0 to 15 point radii

plt.scatter(x, y, s=area, alpha=0.5)
plt.title("Healthy and happy?")
plt.xlabel("Happiness score")
plt.ylabel("Healthy Life Expectancy")
plt.show()
```
